```python
#Importing necessary libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
#Reading the obesity levels data:
data = pd.read_csv("obesity_data.csv")
data.head()
```

```python
#Describing features of the "obesity_data":

#The describe function will calculate the mean, standard deviation, and values necessary to create a box and whisker's plot
#Important: this function will only take into consideration the columns which have numerical values only.

data.describe()
```

```python
#Visual representation of spread of population characteristics:

#These characteristics are age, height, weight and family history.
#In this visualization, we disregard the gender of the individual as this is an overall spread of data.

plt.figure(figsize = (20, 4))
plt.subplot(1, 4, 1)
sns.countplot(data = data, x = "Age", palette = "magma", edgecolor = "black")
plt.subplot(1, 4, 2)
sns.countplot(data = data, x = "Weight", palette = "flare", edgecolor = "black")
plt.subplot(1, 4, 3)
sns.countplot(data = data, x = "Height", palette = "crest", edgecolor = "black")
plt.subplot(1, 4, 4)
sns.countplot(data = data, x = "family_history_with_overweight", palette = "rocket_r", edgecolor = "black")
plt.show()
```

```python
#Visualization of unhealthy eating habits (food and beverage consumption):

#These eating habits are FAVC, CAEC and CALC.
#This visualization is once more an overall model, which disregards the population's gender.

plt.figure(figsize = (20, 4))
plt.subplot(1, 3, 1)
sns.countplot(data = data, x = "FAVC", palette = "magma", edgecolor = "black")
plt.subplot(1, 3, 2)
sns.countplot(data = data, x = "CAEC", palette = "flag", edgecolor = "black")
plt.subplot(1, 3, 3)
sns.countplot(data = data, x = "CALC", palette = "crest", edgecolor = "black")
plt.show()
```

```python
#Visual representation of daily habits of population:

#These daily habits include SMOKE, MTRANS and FAF.
#Once more, this specific visualization will be a generalization of both male and female.

plt.figure(figsize = (20, 4))
plt.subplot(1, 2, 1)
sns.countplot(data = data, x = "MTRANS", palette = "flare", edgecolor = "black")
plt.subplot(1, 2, 2)
sns.countplot(data = data, x = "SMOKE", palette = "rocket_r", edgecolor = "black")
plt.show()
```

```python
#Visualizing the obesity level of the population:

#This demonstration will use the column of the CSV file known as NObeyesdad.
#Again, this will consider only the overall population with no segregation of genders.

sns.countplot(x = "NObeyesdad", data = data, palette = "flare", edgecolor = "black")
plt.title('NObeyesdad', weight='bold')
plt.xticks(rotation=90)
# changing the font size
sns.set(font_scale=1)

```

```python
#Making non-numerical values into numerical values following a matrix method:
ordinal_mappings = {
    'CALC': {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3},
    'CAEC': {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3},
    'NObeyesdad': {'Insufficient_Weight': 0, 'Normal_Weight': 1, 'Overweight_Level_I': 2, 
                   'Overweight_Level_II': 3, 'Obesity_Type_I': 4, 'Obesity_Type_II': 5, 'Obesity_Type_III': 6}
}

# Apply label encoding for ordinal columns
for col, mapping in ordinal_mappings.items():
    data[col] = data[col].map(mapping)

# Apply label encoding for binary columns
binary_mappings = {'yes': 1, 'no': 0}
binary_columns = ['FAVC', 'SCC', 'SMOKE', 'family_history_with_overweight']
for col in binary_columns:
    data[col] = data[col].map(binary_mappings)

# One-hot encoding for nominal columns
data = pd.get_dummies(data, columns=['Gender', 'MTRANS'], drop_first=True)

# Display the transformed DataFrame

# Calculate the correlation matrix
correlation_matrix = data.corr(method='pearson')
correlation_matrix_rounded = round(correlation_matrix, 3)

# Display the rounded correlation matrix
print(correlation_matrix_rounded)
correlation_series = correlation_matrix.unstack().sort_values(kind="quicksort")

# Remove self-correlations
correlation_series = correlation_series[correlation_series != 1]

# Display the highest positive and highest negative correlations
highest_positive_corr = correlation_series.tail(1)
highest_negative_corr = correlation_series.head(1)

print("Highest Positive Correlation:\n", highest_positive_corr)
print("Highest Negative Correlation:\n", highest_negative_corr)
```

```python
mask = np.triu(np.ones_like(correlation_matrix_rounded)) 
plt.figure(figsize=(16, 6))
sns.heatmap(correlation_matrix_rounded, vmin=-1, vmax=1, annot=True, mask = mask)
```
